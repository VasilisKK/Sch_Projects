```r
get_data=function(bike_train,remove_outliers=FALSE,rename=FALSE,keep_both=TRUE) {
   

      #-------------------------- remove outliers ----------------------------
 if (remove_outliers==TRUE) {
# Διαγραφή ακραίων τιμών για windspeed
    Q <- quantile(bike_train$windspeed, probs = c(.25,.75), na.rm = FALSE)
    iqr <- IQR(bike_train$windspeed)
    up <- Q[2] + 1.5 * iqr
    low <- Q[1] - 1.5 * iqr
    bike_train <- subset(bike_train, bike_train$windspeed > (Q[1]-1.5*iqr) & bike_train$windspeed < (Q[2]+1.5*iqr))
# Όμοια και για την μεταβλητή hum
    Q <- quantile(bike_train$hum, probs = c(.25,.75), na.rm = FALSE)
    iqr <- IQR(bike_train$hum)
    up <- Q[2] + 1.5 * iqr
    low <- Q[1] - 1.5 * iqr
    bike_train <- subset(bike_train, bike_train$hum > (Q[1]-1.5*iqr) & bike_train$hum < (Q[2]+1.5*iqr)) }
    
 # Αλλαγή ονομάτος για τις παρακάτω μεταβλητές.
 
 if( rename==TRUE) { 
        bike_train <-bike_train %>% rename("workingday_NO_WORKING_DAY"="workingday_NO WORKING DAY",
                "workingday_WORKING_DAY"="workingday_WORKING DAY", "holiday_NO_HOLIDAY"="holiday_NO HOLIDAY")
 }
# feature engineering 
 if(keep_both==TRUE){
      bike_train <- dummy_cols(bike_train,select_columns=c("holiday","weekday","weathersit"),
                          remove_selected_columns=TRUE) 
      bike_train['season1']= (bike_train$season=="FALL") *201 +(bike_train$season=="WINTER")*156+
       (bike_train$season=="SPRING")*76 +(bike_train$season=="SUMMER")*164
      bike_train['working_dayz']=150*(bike_train$workingday=="WORKING DAY")+121*(bike_train$workingday=="NO WORKING DAY")
      
      bike_train=bike_train[,!(names(bike_train) %in% c("season","workingday"))] }
    
 if(keep_both==FALSE){
      bike_train <- dummy_cols(bike_train,select_columns=c("holiday","weekday","workingday","weathersit"),
                          remove_selected_columns=TRUE)
      bike_train['season1']= (bike_train$season=="FALL") *201 +(bike_train$season=="WINTER")*156+
       (bike_train$season=="SPRING")*76 +(bike_train$season=="SUMMER")*164
      bike_train=bike_train[,!(names(bike_train) %in% c("season"))] }
     
 


 return (bike_train)
  
 

}
```

```r
new_pred<-function (pred) {

    predt <- c()
    for (var in 1:length(pred)) {
        if (pred[var] <= 0) {
            predt[var] <- 1
        }
        if (pred[var] >= 0) {
            predt[var] <- pred[var]
        }
    }
    return(predt)
}

new_pred(c(1,-1,2,-2))
```

```r
rmse_cust<-function (pred, cnt) {

    return(RMSE(new_pred(pred = pred), cnt))
}
```

```r
get_sub=function(bike_test,model,keep_both=TRUE){
    ID=bike_test$ID

    head(bike_test)
    summary(bike_test)
    if (keep_both==TRUE) {
        bike_test <- dummy_cols(bike_test,select_columns=c("holiday","weekday","weathersit"),remove_selected_columns=TRUE)
        bike_test['season1']= (bike_test$season=="FALL") *201 +(bike_test$season=="WINTER")*156+
        (bike_test$season=="SPRING")*76 +(bike_test$season=="SUMMER")*164
       
        bike_test['working_dayz']=150*(bike_test$workingday=="WORKING DAY")+121*(bike_test$workingday=="NO WORKING DAY")
        bike_test=bike_test[,!(names(bike_test) %in% c("season","workingday"))] 
        head(bike_test)}
    
    if (keep_both==FALSE){
         bike_test <- dummy_cols(bike_test,select_columns=c("holiday","weekday","workingday","weathersit"),remove_selected_columns=TRUE)
         bike_test['season1']= (bike_test$season=="FALL") *201 +(bike_test$season=="WINTER")*156+
        (bike_test$season=="SPRING")*76 +(bike_test$season=="SUMMER")*164
         bike_test=bike_test[,!(names(bike_test) %in% c("season"))] 
        
        
        
       }
    
    bike_test=bike_test[,!(names(bike_test) %in% c("ID"))]
    
    pred_cnt=predict(model,bike_test)
    cnt=new_pred(pred_cnt)
    
   
    
    subm=cbind(ID,cnt)
    return (subm)
    
    
    
    
}
```

\section{Stacking}
παρακάτω ακολουθεί το πρώτο μοντέλο που φτιάξαμε με stacking .Με το μοντέλο αυτο κάναμε την καλυτέρη μας υποβόλη.

```r
bike=read.csv('bike_train.csv')
bike=get_data(bike,remove_outliers = TRUE,keep_both = FALSE)


set.seed(100)
# Χωρίζονται τα δεδομένα σε 80/20
index <- createDataPartition(bike$cnt, p = 0.8, list = FALSE)
trainData <- bike[index,]
testData <- bike[-index,]


```

```r
trainControl1 <- trainControl(method="repeatedcv", number=5, repeats=5, savePredictions='final') 



models <- caretList(
  cnt~., data=trainData,
  trControl = trainControl1,
  
 
    
    tuneList = list( rf=caretModelSpec(method='rf'),gb=caretModelSpec(method='gbm',tuneGrid=
                        expand.grid(interaction.depth=c(1,5,9),
                        n.trees=c(500,750,1000,1200), shrinkage = c(0.1,0.01), n.minobsinnode=20)),
                     xgbT=caretModelSpec(method='xgbTree',
                     tuneGrid=expand.grid(nrounds = 1200,eta = 0.01,
                                max_depth =8,gamma=1,colsample_bytree=1,min_child_weight=1,subsample=1))))
```

```r
#load('my_models.RData')
```

```r
models
```

```r
stackControl <- trainControl(method="repeatedcv", number=5, repeats=3, savePredictions='final', ) 
set.seed(7) 
stack.lm <- caretStack(models, method="lm", trControl=stackControl) 
print(stack.lm)
```

```r
preds_lm=predict(stack.lm,testData)
accuracy(preds_lm,testData$cnt)
rmse_cust(preds_lm,testData$cnt)
```

```r
stackControl <- trainControl(method="repeatedcv", number=5, repeats=3, savePredictions='final', ) 
set.seed(7) 
stack.ridge <- caretStack(models, method="ridge", trControl=stackControl,tuneLength=10) 
print(stack.lm)
```

```r
preds_ridge=predict(stack.ridge,testData)
accuracy(preds_ridge,testData$cnt)
rmse_cust(preds_ridge,testData$cnt)
```
